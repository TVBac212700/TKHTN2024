```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import MultipleLocator
import csv
def RdFile(NameFile, Type = int):
    with open(NameFile, newline='', encoding='utf-8') as file:
        Reader = csv.reader(file)
        Columns = next(Reader)
        Data = [row for row in Reader]
    Value = np.array(Data, dtype=Type)
    DataFrame = {Col: Value[:, i] for i, Col in enumerate(Columns)}
    return DataFrame
TimeUART = RdFile("CaculateTime.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 51)
plt.ylim(0, 4500)
plt.plot(TimeUART['N'], TimeUART['T'], marker='o', linestyle='-', color='b')
plt.title('Thời gian truyền tin qua UART với Baurate 115200', fontsize=14)
plt.xlabel('Số ký tự được truyền qua', fontsize=12)
plt.ylabel('Thời gian cần thiết để truyền (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
TimeLCD = RdFile("CaculateLCD.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 17)
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.gca().yaxis.set_major_locator(MultipleLocator(2000))
plt.ylim(2000, 22000)
plt.plot(TimeLCD['N'], TimeLCD['T'], marker='o', linestyle='-', color='r')
plt.title('Thời gian in ra màn hình LCD', fontsize=14)
plt.xlabel('Số ký tự được in lên màn hình', fontsize=12)
plt.ylabel('Thời gian cần thiết để in (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
