```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import MultipleLocator
import csv
def RdFile(NameFile, Type = int):
    with open(NameFile, newline='', encoding='utf-8') as file:
        Reader = csv.reader(file)
        Columns = next(Reader)
        Data = [row for row in Reader]
    Value = np.array(Data, dtype=Type)
    DataFrame = {Col: Value[:, i] for i, Col in enumerate(Columns)}
    return DataFrame
TimeUART = RdFile("CaculateTime.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 51)
plt.ylim(0, 4500)
plt.plot(TimeUART['N'], TimeUART['T'], marker='o', linestyle='-', color='b')
plt.title('Thi gian truyn tin qua UART vi Baurate 115200', fontsize=14)
plt.xlabel('S k t c truyn qua', fontsize=12)
plt.ylabel('Thi gian cn thit  truyn (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
TimeLCD = RdFile("CaculateLCD.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 17)
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.gca().yaxis.set_major_locator(MultipleLocator(2000))
plt.ylim(2000, 22000)
plt.plot(TimeLCD['N'], TimeLCD['T'], marker='o', linestyle='-', color='r')
plt.title('Thi gian in ra mn hnh LCD', fontsize=14)
plt.xlabel('S k t c in ln mn hnh', fontsize=12)
plt.ylabel('Thi gian cn thit  in (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
TimeLCDTask = RdFile("CaculateLCDTask.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 160)
# plt.gca().xaxis.set_major_locator(MultipleLocator(1))
# plt.gca().yaxis.set_major_locator(MultipleLocator(2000))
plt.ylim(20000, 45000)
plt.plot(TimeLCDTask['T'], marker='o', linestyle='-', color='r')
plt.title('Thi gian Task LCD', fontsize=14)
plt.ylabel('Thi gian cn thit  in (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
TimeLCDTask = RdFile("CaculateRXTask.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 160)
plt.plot(TimeLCDTask['T'], marker='o', linestyle='-', color='r')
plt.title('Thi gian Task RX', fontsize=14)
plt.ylabel('Thi gian cn thit  in (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()

TimeLCDTask = RdFile("CaculateTXTask.csv")
plt.figure(figsize=(20,10))
plt.xlim(0, 120)
plt.ylim(0, 3500)
plt.plot(TimeLCDTask['T'], marker='o', linestyle='-', color='r')
plt.title('Thi gian Task RX', fontsize=14)
plt.ylabel('Thi gian cn thit  in (uS)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
